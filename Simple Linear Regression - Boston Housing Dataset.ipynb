### Import Required Packages

```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

### Load Dataset into dataframes

The **datasets** module in **sklearn** contains numerous datasets. From this module, the **Boston House Prices** dataset is loaded into the variable named **house_data**.

```python
house_data = datasets.load_boston()
```

The data and features of this dataset is loaded into a **pandas** dataframe named **house**.

```python
house = pd.DataFrame(house_data.data, columns=house_data.feature_names)
house.head()
```

The target data and target feature of the dataset is loaded into another dataframe named **target**

```python
target = pd.DataFrame(house_data.target, columns=['MEDV'])
target.head()
```

### Visualize Data using scatter plot

The scatter plot shown below compares the independent variable **RM**(Avg no. of rooms per dwelling) of the dataset with the dependent variable **MEDV**(Median value of owner-occupied homes in $1000's)

```python
plt.rcParams['figure.figsize'] = 10,5
plt.scatter(house['RM'], target['MEDV'])
plt.xlabel('Avg no. of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes in $1000')
plt.show()
```

The above plot shows a likely increasing trend between the 2 variables defined previously. 

### Split data into train and test sets

The following code segment splits the data and target dataframes into training and testing dataframes.

```python
x_train = np.array(house['RM'][:450]).reshape(-1,1)
y_train = np.array(target['MEDV'][:450]).reshape(-1,1)

x_test = np.array(house['RM'][450:]).reshape(-1,1)
y_test = np.array(target['MEDV'][450:]).reshape(-1,1)
```

### Fit a Linear Regression model to the training data

Here the **Liner Regression** model is instantiated using the **LinerRegression()** module in **sklearn**.<br>
Also the model is fit to our training data.

```python
reg = LinearRegression().fit(x_train, y_train)
```

### Evaluate fit pictorially

The following scatter plot shows the performance of the model(pictorially) with testing and training data.

```python
plt.rcParams['figure.figsize'] = 20,6
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(x_test, y_test, color='gray')
ax1.plot(x_test, reg.predict(x_test))
ax1.set_xlabel('Avg no. of rooms per dwelling')
ax1.set_ylabel('Median value of owner-occupied homes in $1000')
ax1.set_title('Performance with testing data')

ax2.scatter(x_train, y_train, color='gray')
ax2.plot(x_train, reg.predict(x_train))
ax2.set_xlabel('Avg no. of rooms per dwelling')
ax2.set_ylabel('Median value of owner-occupied homes in $1000')
ax2.set_title('Performance with training data')
```

### Evaluate Fit Statistically

Using the **metrics** module in **sklearn** library, we can compare the target test data with the predicted data.  

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, reg.predict(x_test)))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, reg.predict(x_test)))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, reg.predict(x_test))))
```
